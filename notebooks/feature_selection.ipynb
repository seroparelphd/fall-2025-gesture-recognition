# Feature Selection Report


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.report_utils import save_notebook_artifacts
```

```python
root = Path('..')
data_dir = root / 'data' / 'processed'
results_dir = root / 'results'
feature_importance_df = pd.read_csv(data_dir / 'feature_importance_ranking.csv')
feature_selection_df = pd.read_csv(results_dir / 'feature_selection.csv')
train_selected = pd.read_csv(data_dir / 'train_calib_selected.csv')
test_selected = pd.read_csv(data_dir / 'test_holdout_selected.csv')
print('feature_importance_df:', feature_importance_df.shape)
print('feature_selection_df:', feature_selection_df.shape)
print('train_selected:', train_selected.shape)
print('test_selected:', test_selected.shape)
```

```python
top_20 = feature_importance_df.head(20)
plt.figure(figsize=(8, 6))
sns.barplot(data=top_20, x='importance', y='feature', color='steelblue')
plt.title('Top 20 Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```
